```python
def printlist(temp):
    example = temp
    case_list_righ = str(example).replace('u\'','\'')
    print case_list_righ.decode("unicode-escape")
```

```python
# a set to record all the books in the corpus
from hanziconv import HanziConv

books_set = set()
count = 0
for label in v.labels[:]:
    if label.endswith('.txt'):
        label = label.split('/')
        label = '/'.join(label[1:-1])
        label = HanziConv.toSimplified(label)
        books_set.add(label)
    else:
        label = HanziConv.toSimplified(label)
        book_set.add(label)
        count += 1
printlist(books_set)
print len(books_set)
print count
```

```python
## load the books in the dynastyofbookv2.csv
import csv
books_set_in_csv = set()
with open('dynastyofbookv2.csv','r') as f:
    f_csv = csv.reader(f)
    for row in f_csv:
        book_path = row[0].decode('utf-8')    ## '『史部』/载记/五国故事'
        book_path = book_path.strip(',')
        book_path = HanziConv.toSimplified(book_path)
        books_set_in_csv.add(book_path)
printlist(books_set_in_csv)
print len(books_set_in_csv)
```

```python
### 统计得到CSV文件中书的数量远多余数据集
miss_book_in_corpus = []
for book in books_set:
    if book not in books_set_in_csv:
        miss_book_in_corpus.append(book)
printlist(miss_book_in_corpus)
print len(miss_book_in_corpus)
```

```python
### 统计CSV文件中的哪些书没有出现在数据集中
redundant_book_in_csv = []
for book in books_set_in_csv:
    if book not in books_set:
        redundant_book_in_csv.append(book)
printlist(redundant_book_in_csv)
print len(redundant_book_in_csv)
```
